# Machine Learning
__We will now use some machine learning models to see how these features of interest correlate to the presence of diabetes/prediabetes.__

### Since we are interested in determining whether someone will have diabetes, we will use classifier models to analyze our dataset.

### Lets first try using a KNN model

### We will first split our features between health and non-health related features

```python
# Seperate features and target between health features and non-health features
health_features = df[["HighBP","HighChol","HeartDiseaseorAttack", "GenHlth", "PhysHlth", "BMI", "Age"]]
non_health_features = df[["Education", "Income"]]
target = df["Diabetes_binary"]

# Lists to keep track of accuracies for both health and non-health models
health_acc_list = []
non_health_acc_list = []
```

### Spliting our health and non-health data into training and testing sets

```python
# Train_test_split for health features
x_train_hlth, x_test_hlth, y_train_hlth, y_test_hlth = train_test_split(health_features,target,test_size=0.2,random_state=42)

# Train_test_split for non-health features
x_train_non_hlth, x_test_non_hlth, y_train_non_hlth, y_test_non_hlth = train_test_split(non_health_features,target,test_size=0.2,random_state=42)
```

### Create and fit KNN models for health and non-health features

```python
# Create KNN model for health features
knn_hlth = KNeighborsClassifier()

# Fit the KNN model
knn_hlth.fit(x_train_hlth, y_train_hlth)

# Store training and testing predictions
knn_hlth_pred_train = knn_hlth.predict(x_train_hlth)
knn_hlth_pred_test = knn_hlth.predict(x_test_hlth)

# Print training and testing accuracy
print("Test Accuracy for Health Features: ", accuracy_score(y_test_hlth, knn_hlth_pred_test))
print("Training Accuracy for Health Features: ", accuracy_score(y_train_hlth, knn_hlth_pred_train))

health_acc_list.append(accuracy_score(y_test_hlth, knn_hlth_pred_test))
```

```python
# Create KNN model for non-health features
knn_non_hlth = KNeighborsClassifier()

# Fit the KNN model
knn_non_hlth.fit(x_train_non_hlth, y_train_non_hlth)

# Store training and test set predictions
knn_non_hlth_pred_train = knn_non_hlth.predict(x_train_non_hlth)
knn_non_hlth_pred_test = knn_non_hlth.predict(x_test_non_hlth)

# Print training and testing accuracy
print("Test Accuracy for Non-Health Features: ", accuracy_score(y_test_non_hlth, knn_non_hlth_pred_test))
print("Training Accuracy for Non-Health Features: ", accuracy_score(y_train_non_hlth, knn_non_hlth_pred_train))

non_health_acc_list.append(accuracy_score(y_test_non_hlth, knn_non_hlth_pred_test))
```

### Analysis

Based on the training and testing accuracies, it seems that the KNN model trained on health features performs much better than models trained on non health features. The testing and training accuracies for the health feature KNN model were around 71% and 77% while that of the non health feature model were around 49% and 49%.

This suggests that health features are a better predictor for diabetes than non health features. That is factors such as high blood pressure and cholesterol are a stronger indication of diabetes rather than education and income. 

However, this might also suggest underfitting for the non-health model due to the low training and testing accuracies. This may be explained since we only used 2 non-health features to predict diabetes rather than the 7 we used for health-features, which may have led the model to be unable to capture the true trends of the data. Additionally, through our heatmap, we showed that our dataset is biased towards people of high education and income, which may cause our model to be unable to translate its predictions of other data sets.

### Lets investigate why our non-health model is performing worse than our health model through an error analysis

### Error Analysis

```python
# Confusion matrix for health features
conf_matrix_hlth = confusion_matrix(y_test_hlth, knn_hlth_pred_test)

# Plotting the confusion matrix
sns.heatmap(conf_matrix_hlth, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Health Feature KNN Model')
plt.show()

# Confusion matrix for non features
conf_matrix_non_hlth = confusion_matrix(y_test_non_hlth, knn_non_hlth_pred_test)

# Plotting the confusion matrix
sns.heatmap(conf_matrix_non_hlth, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Non Health Feature KNN Model')
plt.show()
```

Looking at the confusion matrix for the health feature model, we can see that our KNN model trained using health features a high true positive and true negative rate, which is desirable. 

However, looking at the confusion matrix for the non-health feature model, we can see that our KNN model trained using non-health features also has a high true positive rate, but also a high false positive rate and a low true negative rate. This means that when the patient doesn't have diabetes, our model is more likely to incorrectly predict that they do have diabetes. Thus, our model has low specificity.

We hypothesize the reason for this low specificity is due to the feature imbalances in our dataset. As seen in our density heatmap for education and income, a large portion of our dataset consists of people of both high education and income. Thus, for cases where someone doesn't have high education or income, our model will be less able to predict whether they have diabetes. This is especially true for KNN models, since they make their predictions based on the majority of the neighbors, causing feature imbalance to have a large impact on the accuracy. Additionally, this low specificity may also be explained by underfitting, as our KNN model for non-health features only uses 2 features which may not be enough to fully capture the trends of the data.


### Let's try using some more classifier models to see if we can get better results

### Decision Tree Model

```python
# Decision Tree for health features

# Keep track of depth that produces tree with highest testing accuracy
best_depth_hlth = 1
best_train_hlth_acc = 0
best_test_hlth_acc = 0 

for k in range(1, 100):
    tree = DecisionTreeClassifier(max_depth = k)
    tree.fit(x_train_hlth, y_train_hlth)
    
    tree_pred_train = tree.predict(x_train_hlth)
    tree_pred_test = tree.predict(x_test_hlth)
    
    tree_train_acc = accuracy_score(y_train_hlth, tree_pred_train)
    tree_test_acc = accuracy_score(y_test_hlth, tree_pred_test)
    
    
    if tree_test_acc > best_test_hlth_acc:
        best_test_hlth_acc = tree_test_acc
        best_train_hlth_acc = tree_train_acc
        best_depth_hlth = k


print("Best depth:", best_depth_hlth)
print("Best training accuracy for health features:", best_train_hlth_acc)
print("Best testing accuracy for health features:", best_test_hlth_acc)

health_acc_list.append(best_test_hlth_acc)
```

```python
# Decision Tree for non-health features

# Keep track of depth that produces tree with highest testing accuracy
best_depth_non_hlth = 1
best_train_non_hlth_acc = 0
best_test_non_hlth_acc = 0 

for k in range(1, 100):
    tree = DecisionTreeClassifier(max_depth = k)
    tree.fit(x_train_non_hlth, y_train_non_hlth)
    
    tree_pred_train = tree.predict(x_train_non_hlth)
    tree_pred_test = tree.predict(x_test_non_hlth)
    
    tree_train_acc = accuracy_score(y_train_non_hlth, tree_pred_train)
    tree_test_acc = accuracy_score(y_test_non_hlth, tree_pred_test)
    
    if tree_test_acc > best_test_non_hlth_acc:
        best_test_non_hlth_acc = tree_test_acc
        best_train_non_hlth_acc = tree_train_acc
        best_depth_non_hlth = k

print("Best depth:", best_depth_non_hlth)
print("Best training accuracy for health features:", best_train_non_hlth_acc)
print("Best testing accuracy for non-health features:", best_test_non_hlth_acc)

non_health_acc_list.append(best_test_non_hlth_acc)
```

### Analysis

For the health features, the decision tree classifier performed slightly better than the KNN model with training and testing accuracies of 75% and 74%.
For the non-health features, decision tree classifier performed noticably better than the KNN model with training and testing accuracies of 60% and 60%.

We hypothesize that this increase in performance can be explained by how decision trees are better at interpreting from a mixture of feature data types than KNN models, leading to higher accuracies. These results suggest that decision trees are better at capturing the nuances of this dataset, helping combat the underfitting we saw earlier with the KNN model.

### Lets see if a Random Forest performs any better

```python
# Create a Random Forest model for health features
forest_hlth = RandomForestClassifier()

# Fit the Random Forest model
forest_hlth.fit(x_train_hlth, y_train_hlth)

# Store the training and testing predictions
forest_hlth_pred_train = forest_hlth.predict(x_train_hlth)
forest_hlth_pred_test = forest_hlth.predict(x_test_hlth)

# Print the testing and training accuracy
print("Test Accuracy: ", accuracy_score(y_test_hlth, forest_hlth_pred_test))
print("Training Accuracy: ", accuracy_score(y_train_hlth,forest_hlth_pred_train))

health_acc_list.append(accuracy_score(y_test_hlth, forest_hlth_pred_test))
```

```python
# Create a Random Forest model for non-health features
forest_non_hlth = RandomForestClassifier()

# Fit the Random Forest model
forest_non_hlth.fit(x_train_non_hlth, y_train_non_hlth)

# Store the training and testing predictions
forest_non_hlth_pred_train = forest_non_hlth.predict(x_train_non_hlth)
forest_non_hlth_pred_test = forest_non_hlth.predict(x_test_non_hlth)

# Print the testing and training accuracy
print("Test Accuracy: ", accuracy_score(y_test_non_hlth, forest_non_hlth_pred_test))
print("Training Accuracy: ", accuracy_score(y_train_non_hlth,forest_non_hlth_pred_train))

non_health_acc_list.append(accuracy_score(y_test_non_hlth, forest_non_hlth_pred_test))
```

### Analysis

From these accuracies, we can see that for the health features, the random forest classifier actually performed worse than the decision tree classifier. However, for non-health features, the random forest classifier performed similarly to the decision tree classifier.

### Finally, lets use a Logistic Regression model

```python
# Create a logistic regression model for health features
logistic_hlth = LogisticRegression()

# Fit the logistic regression model
logistic_hlth.fit(x_train_hlth, y_train_hlth)

# Store the training and testing predictions
log_hlth_pred_train = logistic_hlth.predict(x_train_hlth)
log_hlth_pred_test = logistic_hlth.predict(x_test_hlth)

# Print the testing and training accuracy
print("Test Accuracy for Health Features: ", accuracy_score(y_test_hlth, log_hlth_pred_test))
print("Training Accuracy for Helath Features: ", accuracy_score(y_train_hlth, log_hlth_pred_train))

health_acc_list.append(accuracy_score(y_test_hlth, log_hlth_pred_test))
```

```python
# Create a logistic regression model for non-health features
logistic_non_hlth = LogisticRegression()

# Fit the logistic regression model
logistic_non_hlth.fit(x_train_non_hlth, y_train_non_hlth)

# Store the training and testing predictions
log_non_hlth_pred_train = logistic_non_hlth.predict(x_train_non_hlth)
log_non_hlth_pred_test = logistic_non_hlth.predict(x_test_non_hlth)

# Print the test and training accuracy
print("Test Accuracy for Health Features: ", accuracy_score(y_test_non_hlth, log_non_hlth_pred_test))
print("Training Accuracy for Health Features: ", accuracy_score(y_train_non_hlth, log_non_hlth_pred_train))

non_health_acc_list.append(accuracy_score(y_test_non_hlth, log_non_hlth_pred_test))
```

### Analysis

From these accuracies, we can see that the logistic regression classifier performed similarly compared to the decision tree and random forest classifier, but still performed better than the KNN model for both health and non-health features.

### Finally, lets compare testing accuracies for both health and non-health models through bar charts

```python
models = ["KNN", "Decision Tree", "Random Forest", "Logistic Regression"]

fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Bar plot for health models
sns.barplot(x=models, y=health_acc_list, ax=axs[0])
axs[0].set_ylabel("Accuracy", fontsize=15)
axs[0].set_title("Accuracies of Health Models", fontsize=15)

# Bar plot for non-health models
sns.barplot(x=models, y=non_health_acc_list, ax=axs[1])
axs[1].set_xlabel("Models", fontsize=15)
axs[1].set_ylabel("Accuracy", fontsize=15)
axs[1].set_title("Accuracies of Non-Health Models", fontsize=15)

plt.tight_layout()

plt.show()


```
